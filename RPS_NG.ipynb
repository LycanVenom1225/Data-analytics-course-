```python
import getpass
def Play_Game():
  print("Let's play a game of Rock, Paper, Scissors. \n you can only select 1 = Rock, 2 = Paper or 3 = Scicors.")
  prompt= ("player 1 please choose 1,2 or 3 only: ")
  prompt_2= ("player 2 please choose 1,2 or 3 only: ")
  message=""
  message_2=""
  options=["1","2","3"]

  while True:
    message= getpass.getpass(prompt)
    if message in options:
      break
    else:
      print("Please enter a valid option")

  while True:
    message_2= getpass.getpass(prompt_2)
    if message_2 in options:
      break
    else:
        print("Please enter a valid option")
  if message==message_2:
    print("It is a tie, no one wins")
  elif message=="1" and message_2=="2":
    print("Paper beats rock, Player_2 wins!")
  elif message=="2" and message_2=="1":
    print("Paper beats rock, Player_1 wins!")
  elif message=="2" and message_2=="3":
    print("Scissors beats Paper, Player_2 wins!")
  elif message=="3" and message_2=="2":
    print("Scissors beats Paper, Player_1 wins!")
  elif message=="3" and message_2=="1":
    print("Rock beats Scissors, Player_2 wins!")
  elif message=="3" and message_2=="1":
    print("Rock beats Scissors, Player_1 wins!")

while True:
  Play_Game()
  choice= input("Do you want to play again Y/N:").upper()
  if choice!="Y":
    print("GAMEOVER")
    break
```
